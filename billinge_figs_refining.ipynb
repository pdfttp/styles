```python
%matplotlib inline
import os

import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
```

```python
def aSeabornStyle(ticks=False, simonCycle=True):
    sns.reset_orig()

    if ticks==False:
        sns.set_style("whitegrid")
    else:
        sns.set_style("ticks")
        
    sns.set_style({
    'grid.linestyle': '--',
    'axes.linewidth': 1,
    'axes.labelcolor':'black',
    'axes.edgecolor': '0.0',
    'xtick.color':'black',
    'ytick.color': 'black',
    'xtick.direction': 'in',
    'ytick.direction': 'in',
    })
                        
    sns.set_context("notebook", font_scale=2, rc={"lines.linewidth": 1.5})

    mpl.rcParams['font.family']='Arial'
    mpl.rcParams['mathtext.default'] = 'regular'
    
    if simonCycle==True:
        simonCycle = ["#0B3C5D", "#062F4F", "#328CC1", "#D9B310", "#984B43", "#B82601",
                 "#57652A", "#76323F", "#626E60", "#AB987A", "#C09F80", "#b0b0b0ff"]
        sns.set_palette(simonCycle)
    else:
        # Use another SNS preset: deep, muted, bright, pastel, dark, colorblind
        sns.set_palette('deep')
```

```python
def anotherstyle():
    sns.reset_orig()
    plt.style.use(os.path.join('./styles/', 'billinge.mplstyle'))
```

```python
def sinplot(flip=1):
    x = np.linspace(0, 14, 100)
    for i in range(1, 13):
        plt.plot(x, np.sin(x + i * .1) * (5 - i) * flip)
        plt.title('Jazz')
```

```python
def NiPDF1():
    r1, g1, d1, c1 = np.loadtxt('./data/Ni.dat', skiprows=4).T

    fig, ax1 = plt.subplots(1, 1, figsize=[14,7])

    ax1.plot(r1, g1, lw=10)
    ax1.plot(r1, c1, lw=2)
    ax1.plot(r1, d1-5, lw=2)

    ax1.set_ylabel('$G(\AA^{-2})$' , fontsize=20)
    ax1.set_xlabel('$r(\AA)$' , fontsize=20)
    
    ax1.set_title('Jazz')
    ax1.set_xlim(1.5,50)
```
