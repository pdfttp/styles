```python
%matplotlib inline
import os

import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
```

```python
def sohSeabornStyle():
    sns.reset_orig()
    sns.set_style("ticks", {'grid.linestyle': '--',
                            'axes.linewidth': 1,
                            'axes.labelcolor':'black',
                            'axes.edgecolor': '0.0',
                            'xtick.color':'black',
                            'ytick.color': 'black',
                            'xtick.direction': 'in',
                            'ytick.direction': 'in',
                             })
                                 
    sns.set_context("notebook", font_scale=2, rc={"lines.linewidth": 1.5})

    # Fig Fonts, and override MathJax font (note 2.0 default is now Arial)
    mpl.rcParams['font.family']='Arial'
    mpl.rcParams['mathtext.default'] = 'regular'
```

```python
def anotherstyle():
    sns.reset_orig()
    plt.style.use(os.path.join('./styles/', 'billinge.mplstyle'))
```

```python
def sinplot(flip=1):
    x = np.linspace(0, 14, 100)
    for i in range(1, 7):
        plt.plot(x, np.sin(x + i * .5) * (7 - i) * flip)
        plt.title('Jazz')
```
