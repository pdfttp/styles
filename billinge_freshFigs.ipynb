```python
%matplotlib inline
import os

import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
```

```python
def sohamSeabornStyle():
    sns.reset_orig()
    sns.set_style("ticks", {'grid.linestyle': '--',
                            'axes.linewidth': 2.5,
                            'axes.labelcolor':'black',
                            'grid.linewidth': 0.01,
                            'axes.edgecolor': '0.3',
                            'xtick.color':'black',
                            'ytick.color': 'black',
                             })
                                 

    sns.set_style({"xtick.direction": "in","ytick.direction": "in"})
    sns.set_context("notebook", font_scale=2.3, rc={"lines.linewidth": 1.5})

    # Fig Fonts, and override MathJax font
    mpl.rcParams['font.family']='Arial'
    mpl.rcParams['mathtext.default'] = 'regular'
```

```python
def anotherstyle():
    sns.reset_orig()
    plt.style.use(os.path.join('./styles/', 'billinge.mplstyle'))
```

```python
def NiPDF(title=None):
    r1, g1, d1, c1 = np.loadtxt('./data/Ni.dat', skiprows=4).T

    fig, ax1 = plt.subplots(1, 1, figsize=[14,7])

    ax1.plot(r1, g1, lw=10, alpha=0.5)
    ax1.plot(r1, c1, lw=2)
    ax1.plot(r1, d1-5)

    ax1.set_ylabel('$G(\AA^{-2})$' , fontsize=20)
    ax1.set_xlabel('$r(\AA)$' , fontsize=20)
    
    ax1.set_title(title)
    ax1.set_xlim(1.5,50)
    
```

```python
def sinplot(flip=1):
    x = np.linspace(0, 14, 100)
    for i in range(1, 7):
        plt.plot(x, np.sin(x + i * .5) * (7 - i) * flip)
```
