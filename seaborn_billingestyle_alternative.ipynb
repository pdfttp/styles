```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns

#Seaborn wrapped styles and context

sns.set_style("ticks", {'grid.linestyle': '--',
                            'axes.linewidth': 2.5,
                            'axes.labelcolor':'black',
                            'grid.linewidth': 0.01,
                            'axes.edgecolor': '0.3',
                            'xtick.color':'black',
                            'ytick.color': 'black',
                             })
                                 

sns.set_style({"xtick.direction": "in","ytick.direction": "in"})
sns.set_context("notebook", font_scale=2.3, rc={"lines.linewidth": 1.5})

# Fig Fonts, and override MathJax font
mpl.rcParams['font.family']='Arial'
mpl.rcParams['mathtext.default'] = 'regular'
```

### Ex 1

```python
def sinplot(flip=1):
    x = np.linspace(0, 14, 100)
    for i in range(1, 7):
        plt.plot(x, np.sin(x + i * .5) * (7 - i) * flip)
```

```python
sinplot()
```

### Ex 2: real life

```python
r1, g1, d1, c1 = np.loadtxt('./data/Ni.dat', skiprows=4).T

fig, ax1 = plt.subplots(1, 1, figsize=[14,7])

ax1.plot(r1, g1)
ax1.plot(r1, c1)
ax1.plot(r1, d1-5)

ax1.set_ylabel('$G(\AA^{-2})$' , fontsize=20)
ax1.set_xlabel('$r(\AA)$' , fontsize=20)


ax1.set_xlim(1.5,50)
```

### Ex 3: more real life
- Standardized line widths, colors, and figure sizes are rarely useful. 
- Here we take the simple plot from Ex. 2, and adjust it. I prefer the SNS color pallete, from which I make colors more and less opaque manually, and I also adjust line widths and markers manually. 
- When needed, I select specific colors by googling "html color picker" and copy paste them into the color field.
- Using a template style sheet as detailed as the one in this repo is very inconvenient. It means we'd have to override the defaults for special situations, where the majority of situations are special.

```python
r1, g1, d1, c1 = np.loadtxt('./data/Ni.dat', skiprows=4).T

fig, ax1 = plt.subplots(1, 1, figsize=[14,7])

ax1.plot(r1, g1, lw=10, color='#307cb2', alpha=0.5)
ax1.plot(r1, c1, lw=2, color='#13a80b')
ax1.plot(r1, d1-5, color='r')

ax1.set_ylabel('$G(\AA^{-2})$' , fontsize=20)
ax1.set_xlabel('$r(\AA)$' , fontsize=20)


ax1.set_xlim(1.5,50)
```
